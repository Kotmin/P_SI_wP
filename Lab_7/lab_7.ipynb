```python
from keras.layers import Conv2D, Flatten, Dense, AveragePooling2D, MaxPooling2D
from keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from keras.datasets import mnist
import numpy as np
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import  GridSearchCV
train, test = mnist.load_data()
```

```python
X_train, y_train = train[0], train[1]
X_test, y_test = test[0], test[1]
X_train = np.expand_dims(X_train, axis=-1)
X_test = np.expand_dims(X_test, axis=-1)
```

```python
train_indices = np.random.randint(0,X_train.shape[0]-1,  size=int(0.3*(X_train.shape[0])))
test_indices = np.random.randint(0,X_test.shape[0]-1,  size=int(0.3*(X_test.shape[0])))
X_train = X_train[train_indices]
y_train = y_train[train_indices]
X_test = X_test[test_indices]
y_test = y_test[test_indices]
```

```python
def build_model(layer_num=1):
  class_cnt = np.unique(y_train).shape[0]
  filter_cnt = 32
  neuron_cnt = 32
  learning_rate = 0.0001
  act_func = 'relu'
  kernel_size = (3,3)
  model = Sequential()
  conv_rule = 'same'
  model.add(Conv2D(input_shape = X_train.shape[1:],
  filters=filter_cnt,
  kernel_size = kernel_size,
  padding = conv_rule, activation = act_func)) # warstwa wejsciowa (input)
  for i in range(layer_num):
    model.add(Conv2D(  filters=filter_cnt,
    kernel_size = kernel_size,
    padding = conv_rule, activation = act_func))
  model.add(Flatten()) # splaszczanie
  model.add(Dense(class_cnt, activation='softmax')) # wyjsciowa
  model.compile(optimizer=Adam(learning_rate),
                loss='SparseCategoricalCrossentropy',
                metrics='accuracy')
  return model
```

```python
keras_classifier=KerasClassifier(build_model)
param_distribs={
    'layer_num': [1,2,3]
    }
num_epochs=5
grid_search_cv=GridSearchCV(keras_classifier, param_distribs, cv=5)
grid_search_cv.fit(X_train, y_train, epochs=num_epochs)
best_params_from_grid=grid_search_cv.best_params_ 
```
